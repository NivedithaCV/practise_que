```python
import numpy as np
import matplotlib.pyplot as plt
def func(mean,cov):
    x,y= np.random.multivariate_normal(mean,cov,100).T
    return(x,y)
def plot(x,y):
    plt.scatter(x,y)
    plt.show()
x1,y1=func([0,0],[[1,0],[0,1]])
```

```python
plot(x1,y1)
```

```python
x2,y2=func([-1,1],[[1,0],[0,1]])
```

```python
plot(x2,y2)
```

```python
x3,y3=func([0,0],[[2,0],[0,2]])
```

```python
plot(x3,y3)
```

```python
x4,y4=func([0,0],[[1,0.5],[0.5,1]])
```

```python
plot(x4,y4)
#shift y=x
```

```python
x5,y5=func([0,0],[[1,-0.5],[-0.5,1]])
```

```python
plot(x5,y5)
#shift y=-x
```
